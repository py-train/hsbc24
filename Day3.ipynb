```python
employees_dict = {}
key_index = 1
lines = employees.strip().splitlines()
fieldnames = lines.pop(0).split(',')
fieldnames.pop(key_index)

for line in lines:
    fields = line.split(',')
    assert fields is not None and len(fields) == len(fieldnames)+1
    name = fields.pop(key_index)
    
    empdict = {}
    for fieldname, field in zip(fieldnames, fields):
        empdict[fieldname] = field

    employees_dict[name] = empdict

employees_dict
```

```python
employees_dict = {}
key_field_name = 'name'
lines = employees.strip().splitlines()
fieldnames = lines.pop(0).split(',')

for line in lines:
    fields = line.split(',')
    assert fields is not None and len(fields) == len(fieldnames)
    
    empdict = {}
    for fieldname, field in zip(fieldnames, fields):
        empdict[fieldname] = field

    key = empdict.pop(key_field_name)
    employees_dict[key] = empdict

employees_dict
```
