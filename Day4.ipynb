```python
ztip = (tip - tip.min())/(tip.max() - tip.min())
ztip
```

```python
ztip.min()
```

```python
ztip.max()
```

```python
tip_pct = tips['tip'] / tips['total_bill'] * 100
```

```python
tip_pct.min(), tip_pct.max()
```

```python
# Avg tip amount for tips less than 10$
tip[tip < 10].mean()
len(tip[tip < 10])
```

```python
# Avg tip amount for tips less than 50%
tip[tip_pct <= 50].mean()
len(tip[tip_pct <= 50])
```
