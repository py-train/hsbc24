```python
mpg_yoc_agg = mpg.groupby(['model_year', 'origin', 'cylinders']).mpg.agg(['min', 'mean', 'max']).unstack()
mpg_yoc_agg
```

```python
# mean mpg for japanese cars from 74-78 with 4 cylinders

res = mpg_yoc_agg.loc[(slice(74,78),'japan'),('mean',4)]
res
```

```python
# mean mpg for japanese cars from 74-78 with 4:6 cylinders

res = mpg_yoc_agg.loc[(slice(74,78),'japan'),('mean',slice(4,6))]
res
```
